# Linear Regression in Python
This notebook demonstrates a simple linear regression analysis using Python to model Salary based on Years of Experience.
The scikit-learn libraries automatically calculate the y-intercept, slope, correlation coefficient, and mean squared error values for the predicted model.

Slope = 8285.292099903527

    Expected increase in salary after 1 year of additional experience.
    
y-intercept = 29203.52267124022

    Expected salary at 0 years of experience.
    
R-squared = 0.785

    Proportion of variation in years of experience that is predictable based on salary (i.e. 78.5% of variation in years of experience is predictable by salary).
    
MSE = 17523844.08

    Average squared difference between the estimated (regression line) and actual (data points) values of the data set.

```python
import pandas as pd
```

Imports the pandas library and gives it a short name (pd). Pandas is a powerful Python library for working with tabular data (like spreadsheets or CSV files).

```python
import matplotlib.pyplot as plt
```

Loads the matplotlib plotting library, which is used to create charts and visualizations.

```python
from sklearn.linear_model import LinearRegression
```

Imports the LinearRegression class from the scikit-learn (sklearn) library. Scikit-learn is a widely used machine learning library.

```python
from sklearn.metrics import mean_squared_error
```

Imports the MeanSquaredError class from the scikit-learn library.

```python
# Load data
data = pd.read_csv("regression_data.csv")
data.columns = data.columns.str.strip()  # strip whitespace
x_col = "YearsExperience"
y_col = "Salary"
```

Reads a CSV file and loads it into a DataFrame, which is like a table in memory that you can work with in Python.

```python
# Select the independent (X) and dependent (y) variables
X = data[[x_col]]
y = data[[y_col]]

# Fit the linear regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)

# Generate predictions
predictions = model.predict(X)

# Calculate regression parameters
slope = model.coef_[0][0]
intercept = model.intercept_[0]
r_squared = model.score(X, y)

# Prepare annotation text
equation = f'y = {slope:.2f}x + {intercept:.2f}'
r_squared_text = f'$R^2$ = {r_squared:.3f}'
annotation_text = f'{equation}\n{r_squared_text}'

# Plot the data and regression line
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='red', label='Actual data')
plt.plot(X, predictions, color='blue', label='Regression line')

# Add the regression equation and R² to the plot
plt.text(
    0.05, 0.95,
    annotation_text,
    transform=plt.gca().transAxes,
    fontsize=12,
    verticalalignment='top',
    bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5')
)

# Final plot formatting
plt.title(f'{y_col} vs {x_col}')
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.legend()
plt.tight_layout()
plt.show()
```

**model.fit(x, y) :** Fits the linear regression model to the data. X is the input (independent variable, must be 2D), y is the target (dependent variable).

**plt.scatter(x, y) :** Creates a scatter plot — red dots on a graph showing the original data points.

**plt.plot(x, y_pred) :** Draws the regression line using predicted values. The blue line is the best-fit line the model calculates.

```python
# Stats
slope = model.coef_[0][0]
intercept = model.intercept_[0]
r_squared = model.score(data[["YearsExperience"]], data[["Salary"]])
mse = mean_squared_error(data["Salary"], model.predict(data[["YearsExperience"]]))
```

```python
print(f"Slope = {slope}")
print(f"y-intercept = {intercept}")
print(f"R-squared = {r_squared:.3f}")
print(f"MSE = {mse:.2f}")
```
